```python
df_des_quan = df.loc[:, ('Description', 'Quantity')]
df_des_quan.to_csv('to_csv_OnlineRetail.csv')
df_des_quan.head()
```

```python
df_large_quantity = df.loc[df.Quantity > 10]
df_large_quantity.to_hdf('to_hdf_OnlineRetail.h5', 'table')
hdf_df = pd.read_hdf('to_hdf_OnlineRetail.h5', key='table')
hdf_df.head()
```

```python
new_df = df.iloc[:1001]
new_df.to_excel('1000_lines_to_excel_OnlineRetail.xlsx', )
new_df.tail()
```

```python
df_quan_in_price = df.loc[1000:2000, ['Quantity', 'InvoiceDate', 'UnitPrice']]
head = df_quan_in_price.head()
tail = df_quan_in_price.tail()
print(head)
print(tail)
df_quan_in_price.info()

df_quan_in_price.to_json('to_json_OnlineRetail.json', orient='columns')
```
