```python
import pandas as pd
```

```python
df = pd.read_csv('FoodPrice_in_Turkey.csv')
df.head()
```

```python
df.info()
```

### Change columns name

```python
df.rename(columns={'Place': 'Vi tri', 'ProductId': 'Ma san pham'}, inplace=True)
df.head()
```

### Add new column

```python
df['New column'] = False
df.head()
```

### Add new discount column

```python
df['Discount'] = pd.Series('10%', index=df.index)
df.head()
```

### Add another discount column

```python
df.insert(df.shape[1], 'Discount but 12%', pd.Series('12%', index=df.index))
df.head()
```

### Add new row

```python
new_df = df.append({
    'Vi tri': 'London',
    'Ma san pham': 999,
    'ProductName': 'cafe',
    'UmId': 10, 'UmName': 'Hello',
    'Month': 7,
    'Year': 2021,
    'Price': 1,
    'New column': True,
    'Discount': '10%',
    'Discount but 12%': None
}, ignore_index=True)
new_df.tail()
```

### Delete a column

```python
del df['New column']
df.head()
```

### Delete with pop

```python
df.pop('Discount but 12%')
```

### Delete with drop

```python
df.drop('Discount', axis=1, inplace=True)
df.head()
```

### Delete multiple column with drop

```python
drop_list = ['Month', 'Year']
df.drop(drop_list, axis=1, inplace=True)
df.head()
```

### Delete a row with drop

```python
df.drop([1, 2, 3], axis=0, inplace=True)
df.head()
```
