```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('shopeep_koreantop_clothing_shop_data.csv')
df
```

```python
df = df[['join_month', 'join_day', 'join_year', 'shop_location', 'rating_good', 'rating_bad', 'rating_normal']]
df
```

```python
#Them cot du lieu moi tu cac cot du lieu kieu so. Su dung cac phep toan + - * /
# Cong thuc rating = rating_good * 2 + rating_normal - rating_bad * 3

df['rating'] = df['rating_good'] * 2 + df['rating_normal'] + df['rating_bad'] * 3
df
```

```python
df.info()
```

```python
# Them cot tu du lieu kieu chuoi
#VD: Them cot date gop tu month, day, year. Format 'month day, year'
df['date'] = df['join_month'].astype(str) + df['join_day'].astype(str) + ',' + df['join_year'].astype(str)
df
```

```python
#Them cot theo dieu kien
#VD: Them cot new, return True neu join_year == 2021, else false
df['new'] = df['join_year'] == 2021
df
```

```python
#Su dung np.where neu co 2 lia chon
#VD: Them cot rate. Good if rating_good >= 50000, else bad
df['rate'] = np.where(df['rating_good'] >= 50000, 'Good', 'Bad')
df
```

```python
# Su dung np.select neu co nhieu hon 2 lua chon
# VD:
# Thêm cột flag tặng cờ cho các cửa hàng, flag nhận các giá trị như sau:
# blue khi rating_good >= 30000 và rating_bad <= 100
# yellow khi 10000 <= rating_good < 30000 và 100 < rating_bad <= 1000
# red khi rating_good < 10000
# black đối với các trường hợp còn lại
conditions = [
    (df['rating_good'] >= 30000) & (df['rating_bad'] <= 100),
    (df['rating_good'] >= 10000) &
    (df['rating_good'] < 30000) &
    (df['rating_bad'] > 100) &
    (df['rating_bad'] <= 1000),
    (df['rating_good'] < 10000)
]
choice_list = ['Blue', 'Yellow', 'Red']

df['flag'] = np.select(condlist=conditions, choicelist=choice_list, default='Black')
df
```
