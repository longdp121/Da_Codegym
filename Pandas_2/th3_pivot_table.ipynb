```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('FoodPrice_in_Turkey.csv')
df.head()
```

```python
df.info()
```

```python
df['Place'].unique()
```

```python
#Caculate price mean of product by year
df_price_mean_by_year = df.pivot_table(values='Price', index='ProductName', columns='Year', aggfunc='mean')
df_price_mean_by_year.head()
```

```python
#Show highest product price by year
df_max_price_by_year = df.pivot_table(values='Price', columns='Year', aggfunc='max')
df_max_price_by_year.head()
```

```python
#Show hightest price of each product
df_max_price_by_product = df.pivot_table(values='Price', index='ProductName', aggfunc='max')
df_max_price_by_product.head()
```

```python
#Show highest, lowest price in each place by month
df_price_in_place_by_month = df.pivot_table(
    values='Price',
    columns='Place',
    index='Month',
    aggfunc={'max', 'min'}
)
df_price_in_place_by_month.head()
```

```python
#Show tootal price and average month selling by product name
year = len(df['Year'].unique())

df_total_price_ave_year_by_productname = df.pivot_table(
    values=['Price', 'Month'],
    columns='Place',
    index='ProductName',
    aggfunc={'Price': np.sum, 'Month': np.mean}
)
df_total_price_ave_year_by_productname
```
