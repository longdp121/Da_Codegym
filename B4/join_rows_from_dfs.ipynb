```python
import pandas as pd
```

```python
df = pd.read_csv('FoodPrice_in_Turkey.csv')
df.head()
```

```python
# Creat new dataframe from first 5000 rows
df_1 = df.loc[:4999, :]
df_1.info()
df_1.tail()
```

```python
# Creat another dataframe which contain the rest
df_2 = df.loc[4999:, :]
df_2.info
df_2.head(10)
```

```python
# Creat a new dataframe contain price details between row 1000 and 2000
df_price_1000_2000 = df.loc[1000:2000, ['ProductId', 'Month', 'Year', 'Price']]
df_price_1000_2000.head()
#df_price_1000_2000.tail()
```

```python
# Concat df_1 and df_2 (by row, axis=0) to creat a new one
new_df = pd.concat([df_1, df_2], axis=0)
new_df.info()
new_df
```

```python
# Concat df_1, df_2 and df_price_1000_2000 together by row (axis=0). Missing vallue will be displayed as NaN
new_new_df = pd.concat([df_1, df_2, df_price_1000_2000], axis=0)
new_new_df.info()
new_new_df
```

```python
# df_1 append df_2 to creat a new one
append_df = df_1.append(df_2)
append_df.info()
append_df
```

```python
# df_1 append df_price_1000_2000 to creat a new one. Missing vallue will be displayed as NaN
new_append_df = df_1.append(df_price_1000_2000)
new_append_df.info()
new_append_df
```
