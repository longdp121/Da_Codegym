```python
import pandas as pd
```

```python
df = pd.read_csv('FoodPrice_in_Turkey.csv')
df.head()
```

```python
df.info()
```

```python
# Delete rows which share productid and keep the last record

new_df = df.drop_duplicates(subset=['ProductId'], keep='last')
new_df.head()
```

```python
# Delete rows whicj share productid, keep last record and reset index (also delete old index)
new_index_df = df.drop_duplicates(subset=['ProductId'], keep='last').reset_index(drop=True)
new_index_df.head()
```

```python
# Creat a new dataframe which contain product information
df_pro = df.loc[:, ['ProductId', 'ProductName', 'UmId', 'UmName']]
df_pro.head()
```

```python
# Creat a new dataframe which contain product price information
df_price = df.loc[:, ['ProductId', 'Place', 'Month', 'Year', 'Price']]
df_price.head()
```

```python
# Creat a new dataframe which contain product anem and price for row 10 to 20
df_price_10_20 = df.loc[10:20, ['ProductId', 'ProductName', 'Price']]
df_price_10_20.head(12)
```

```python
# Merge df_pro and df_price on productid
merge_df = pd.merge(df_pro, df_price, on='ProductId')
merge_df.head()
```

```python
# Concat df_pro and df_price
concat_df = pd.concat([df_pro, df_price], axis=1)
concat_df.head()
```

```python
# Concat df_pro, df_price and df_price_10_20, missing value will be show as NaN
new_nan_df = pd.concat([df_pro, df_price, df_price_10_20], axis=1)
new_nan_df.head(30)
```
