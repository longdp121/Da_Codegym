# Sinh cột dữ liệu dẫn suất
1. Sinh cột dữ liệu tổng hợp từ các cột khác
2. Tách cột dữ liệu từ bảng có sẵn

```python
#1. Sinh cột dữ liệu
df_sale = pd.read_csv('OnlineRetail.csv')
df_sale.head()
```

```python
df_sale.info()
```

```python
#Thêm cột total price = UnitPrice * Quantity
df_sale['TotalPrice'] = df_sale['Quantity'] * df_sale['UnitPrice']
df_sale.head()
```

```python
#Thêm VAT (10%) vào cột TotalPrice
VAT = 0.1
df_sale['IncludeVAT'] = df_sale['TotalPrice'] * (1 + VAT)
df_sale.head()
```

```python
df_sale.describe()
```

```python
#Thêm cột giá trị True, False để phân biệt đơn giá bị âm
df_sale['negative_price'] = df_sale.UnitPrice < 0
df_sale
```

```python
#Tìm bản ghi có giá trị âm
df_sale.loc[df_sale.negative_price == True]
```

```python
#Sử dụng np.where (lệnh where của numpy)
new_arr = np.where(df_sale.UnitPrice < 0, True, False)
print(new_arr)
print(type(new_arr))
```

```python
df_sale['np_check_price'] = new_arr
df_sale.head()
```

```python
temp = []
for i in range(len(new_arr)):
    if new_arr[i] == True:
        temp.append(i)

temp
```

```python
df_sale['np_new_price'] = np.where(df_sale.UnitPrice <= 0, 1000, df_sale.UnitPrice)
df_sale.head()
```

```python
df_sale[df_sale.UnitPrice != df_sale.np_new_price]
```

```python
df_sale[df_sale.np_check_price]
```
