```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler
```

```python
df = pd.read_csv('FoodPrice_in_Turkey.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna()
```

```python
df1 = df.dropna()
df1.info()
```

```python
#Xu ly du lieu ngoai lai o cot Price
sns.boxplot(x=df1['Price'])
```

```python
Q1 = df1['Price'].quantile(0.25)
Q3 = df1['Price'].quantile(0.75)

IQR = Q3 - Q1
IQR
```

```python
#Xac dinh phan tu khong ngoai lai
df2 = df1
df2['outliner'] = ~((df1['Price'] < Q1 - 1.5 * IQR) | (df1['Price'] > Q3 + 1.5 * IQR))
```

```python
df2
```

```python
df2 = df2[df2['outliner'] == True]
df2
```

```python
sns.kdeplot(data=df2['Price'])
```

```python
#Chuan hoa du lieu
df2['Price'].describe()
```

```python
scaler = MinMaxScaler()
df_s = scaler.fit_transform(df2[['Price']])
```
