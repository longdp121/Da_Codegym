```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_boston

boston = load_boston()
data = boston.data
y = boston.target
columns = boston.feature_names
boston_df = pd.DataFrame(data)
boston_df.columns = columns
# data_url = "http://lib.stat.cmu.edu/datasets/boston"
# boston_df = pd.read_csv(data_url, sep='\s+', skiprows=22, header=None)
```

```python
#Xac dinh ngoaij lai bang toan hoc
#Z-Score
from scipy import stats

z = np.abs(stats.zscore(boston_df))
print(type(z))
z
```

```python
threshold = 3
outliners = np.where(z > threshold)
print(outliners)
label_list = z.columns.tolist()
row_index, col_index = outliners
for i in range(0, len(row_index)):
    print(f'Outliners are z[{row_index[i]}][{label_list[col_index[i]]}] = {z.to_numpy()[row_index[i]][col_index[i]]}')
```

```python
boston_df_zscore = boston_df[(z < 3).all(axis=1)]
boston_df_zscore
```

```python
#IQR
Q1 = boston_df.quantile(0.25)
Q3 = boston_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
(boston_df < (Q1 - 1.5 * IQR)) | (boston_df > (Q3 + 3 * IQR))
```

```python
boston_df_iqr = boston_df[~((boston_df < (Q1 - 1.5 * IQR)) | (boston_df > (Q3 + 3 * IQR))).any(axis=1)]
boston_df_iqr
```

```python
boston_df_iqr.loc[boston_df_iqr['DIS'] > 10]
```
